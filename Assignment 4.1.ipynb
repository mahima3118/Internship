```python
soup = BeautifulSoup(page.content)
soup
```

```python
table = soup.find("table", class_="wikitable")

rank_list = []
name_list = []
artist_list = []
views_list = []
upload_date_list = []

for row in table.find_all("tr")[1:]:
    columns = row.find_all("td")
    if len(columns) >= 6:
        rank = columns[0].get_text().strip()
        name = columns[1].get_text().strip()
        artist = columns[2].get_text().strip()
        views = columns[3].get_text().strip()
        upload_date = columns[4].get_text().strip()
        

        rank_list.append(rank)
        name_list.append(name)
        artist_list.append(artist)
        views_list.append(views)
        upload_date_list.append(upload_date)
df = pd.DataFrame({
    "Rank": rank_list,
    "Name": name_list,
    "Artist": artist_list,
    "Views": views_list,
    "Upload Date": upload_date_list
})

# Print the DataFrame
print(df)
```

```python
# method 1
rank_list = []
state_list = []
gsdp_18_19_list = []
gsdp_19_20_list = []
share_18_19_list = []
gdp_billion_list = []

# for odd rows

rows_odd = driver.find_elements(By.CLASS_NAME, "odd")  # help as i am getting output twice so i managed with dropduplicate


for row_odd in rows_odd:
    columns_odd = row_odd.find_elements(By.TAG_NAME, "td")
    rank_odd = columns_odd[0].text.strip()
    state_odd = columns_odd[1].text.strip()
    gsdp_18_19_odd = columns_odd[3].text.strip()
    gsdp_19_20_odd = columns_odd[2].text.strip()
    share_18_19_odd = columns_odd[4].text.strip()
    gdp_billion_odd = columns_odd[5].text.strip()

# for even rows    
    try:
        row_even = row_odd.find_element(By.XPATH, "./following-sibling::tr[@class='even']")
        columns_even = row_even.find_elements(By.TAG_NAME, "td")
        rank_even = columns_even[0].text.strip()
        state_even = columns_even[1].text.strip()
        gsdp_18_19_even = columns_even[3].text.strip()
        gsdp_19_20_even = columns_even[2].text.strip()
        share_18_19_even = columns_even[4].text.strip()
        gdp_billion_even = columns_even[5].text.strip()
    except NoSuchElementException:
# If no corresponding "even" row, use empty values
        rank_even = ""
        state_even = ""
        gsdp_18_19_even = ""
        gsdp_19_20_even = ""
        share_18_19_even = ""
        gdp_billion_even = ""

    
    rank_list.extend([rank_odd, rank_even])
    state_list.extend([state_odd, state_even])
    gsdp_18_19_list.extend([gsdp_18_19_odd, gsdp_18_19_even])
    gsdp_19_20_list.extend([gsdp_19_20_odd, gsdp_19_20_even])
    share_18_19_list.extend([share_18_19_odd, share_18_19_even])
    gdp_billion_list.extend([gdp_billion_odd, gdp_billion_even])


df = pd.DataFrame({
    "Rank": rank_list,
    "State": state_list,
    "GSDP(18-19) - at current prices": gsdp_18_19_list,
    "GSDP(19-20) - at current prices": gsdp_19_20_list,
    "Share(18-19)": share_18_19_list,
    "GDP($ billion)": gdp_billion_list
})


print(df.to_string(index=False, max_rows=None, max_cols=None))

```

```python
repo_urls = []

scrolls = 6  
for _ in range(scrolls):
    driver.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
    time.sleep(2) 

repo_links = driver.find_elements(By.XPATH, '//a[@class="Link"]')

repo_urls.extend([link.get_attribute("href") for link in repo_links])

for url in repo_urls:
    print(url)
```

```python

repository_titles = []
repository_descriptions = []
contributors_counts = []
languages_used = []




for i in repo_urls:
    driver.get(i)
    time.sleep(5)
    
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)  
    
    try:
        titles=driver.find_element(By.XPATH,'//strong[@class="mr-2 flex-self-stretch"]')
        repository_titles.append(titles.text)
    except NoSuchElementException :
        repository_titles.append('not present')
    try:
        descriptions=driver.find_element(By.XPATH,'//p[@class="f4 mb-3 "]')
        repository_descriptions.append(descriptions.text)
    except NoSuchElementException :
        repository_descriptions.append('not present')
    try:
        counts=driver.find_element(By.XPATH,'/html/body/div[1]/div[4]/div/main/turbo-frame/div/div/div/div[2]/div[2]/div/div[4]/div/h2/a/span')
        contributors_counts.append(counts.text)
    except NoSuchElementException :
        contributors_counts.append('not present')
    try:
        languages=driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        languages_used.append(languages.text)
    except NoSuchElementException :
        languages_used.append('not present') 
    
    



data = {
    'Repository_Titles': repository_titles,
    'Repository_Descriptions': repository_descriptions,
    'Languages_Used': languages_used,
    'Contributors_Counts': contributors_counts
}


df = pd.DataFrame(data)


print(df)
    
    
    
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

L
url = "https://www.imdb.com/list/ls095964455/"
response = requests.get(url)


soup = BeautifulSoup(response.text, "html.parser")


names = []
year_spans = []
genres = []
runtimes = []
ratings = []
votes = []


container = soup.find("div", {"class": "lister-list"})


for series in container.find_all("div", {"class": "lister-item"}):
    name = series.find("h3", {"class": "lister-item-header"}).a.text.strip()
    year_span = series.find("span", {"class": "lister-item-year"}).text.strip("()").strip()
    
   
    genre = series.find("span", {"class": "genre"})
    runtime = series.find("span", {"class": "runtime"})
    
    if genre:
        genre = genre.text.strip()
    else:
        genre = "N/A"
    
    if runtime:
        runtime = runtime.text.strip()
    else:
        runtime = "N/A"
    
    rating = series.find("span", {"class": "ipl-rating-star__rating"}).text.strip()
    
    
    vote_element = series.find("span", {"name": "rk"})

    
    vote = vote_element.text.strip() if vote_element else "N/A"

    # Append data to lists
    names.append(name)
    year_spans.append(year_span)
    genres.append(genre)
    runtimes.append(runtime)
    ratings.append(rating)
    votes.append(vote)

# Create a DataFrame from the scraped data
data = {
    "Name": names,
    "Year Span": year_spans,
    "Genre": genres,
    "Run Time": runtimes,
    "Ratings": ratings,
    "Votes": votes,
}

df = pd.DataFrame(data)

# Print the DataFrame
print(df)
```

```python
# get urls
product_urls = []



product_links = driver.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')

product_urls.extend([link.get_attribute("href") for link in product_links])


for url in product_urls:
    print(url)
```

```python
dataset_name = []
dataset_type = []
task = []
attribute_type = []
no_of_instances = []
no_of_attributes = []
year = []

for i in product_urls:
    driver.get(i)
    time.sleep(5)

    
    dataset_name_element = driver.find_element(By.XPATH, '//h1[@class="text-3xl font-semibold text-primary-content"]')
    dataset_type_element = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p')
    task_element = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p')
    attribute_type_element = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p')
    no_of_instances_element = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p')
    no_of_attributes_element = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/p')
    try:
        year_element = driver.find_element(By.XPATH, '//h2[@class="text-sm text-primary-content"]').text.strip()
    except:
        year_element = "NA" 
    
    
    dataset_name.append(dataset_name_element.text.strip())
    dataset_type.append(dataset_type_element.text.strip())
    task.append(task_element.text.strip())
    attribute_type.append(attribute_type_element.text.strip())
    no_of_instances.append(no_of_instances_element.text.strip())
    no_of_attributes.append(no_of_attributes_element.text.strip())
    year.append(year_element)


df = pd.DataFrame({
    "Dataset_Name": dataset_name,
    "Data_Type": dataset_type,
    "Task": task,
    "Attribute_Type": attribute_type,
    "No_of_Instances": no_of_instances,
    "No_of_Attributes": no_of_attributes,
    "Year": year
})

print(df)
```
